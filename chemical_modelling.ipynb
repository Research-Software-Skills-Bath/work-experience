### Task 4: Animating bonds forming and breaking
Behind every 3D graphic on a computer is a hidden list of numbers. In chemistry, one of the most common ways to save a molecule is in an XYZ file format, shown in the below celss between """ and """ .
For each Frame the computer reads three things:
1. The Atom Count: The very first number tells the computer how many atoms to look for.
2. The Name Tag: A line of text describing the frame (the computer ignores this, it is just for humans).
3. The 3D Coordinates: A list of the atoms (O for Oxygen, H for Hydrogen) followed by three numbers which act like a 3D map grid:
    * X (First number): Moves the atom Left (-) or Right (+)
    * Y (Second number): Moves the atom Down (-) or Up (+)
    * Z (Third number): Moves the atom Back (-) or Forward (+)
    
* Change some numbers in the XYZ data, what happens to the animation?

```python
# A flipbook of 3 frames showing: [Reactants] -> [Collision] -> [Products]
reaction_frames = """
3
Frame 1: Reactants heading toward each other
O    0.000    0.000    0.000
H   -1.500    1.000    0.000
H    1.500    1.000    0.000
3
Frame 2: The collision state (atoms breaking apart)
O    0.000    0.000    0.000
H   -0.800    0.500    0.000
H    0.800    0.500    0.000
3
Frame 3: Products (Water molecule formed!)
O    0.000    0.000    0.000
H   -0.757    0.586    0.000
H    0.757    0.586    0.000
"""

# Create the canvas
viewer = py3Dmol.view(width=500, height=400)

# CRITICAL: Tell py3Dmol to read this text as an animated movie trajectory
viewer.addModelsAsFrames(reaction_frames, 'xyz')

# Make the atoms big and clear
viewer.setStyle({}, {'sphere': {'scale': 0.3}, 'stick': {'radius': 0.15}})
viewer.setBackgroundColor('white')
viewer.zoomTo()

# Play the movie! (loop: "forward" plays it over and over)
viewer.animate({'loop': 'forward', 'interval': 500}) 

viewer.show()
```
