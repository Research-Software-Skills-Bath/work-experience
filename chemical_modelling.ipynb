```python
# Initialize viewer
view = py3Dmol.view(width=800, height=400)

# Fetch molecule from PubChem by CID (e.g., Aspirin = CID 2244)
# You can also pass custom SDF/XYZ strings here if you have multiconformer data
view.addModelsAsFrames("cid:2244", "sdf")

# Set display style for the models
view.setStyle({'stick': {'radius': 0.15}, 'sphere': {'scale': 0.3}})

# Start animation
view.animate({'loop': 'forward', 'reps': 1})
view.zoomTo()
display(view.show())
```

```python
# 1. Initialize the grid and pre-load both CIDs at the same time
v = py3Dmol.view(query=['cid:2244','cid:702'], viewergrid=(1,2), width=800, height=400)

# 2. Style the first cell (Water on the left)
v.setStyle({'sphere': {}}, viewer=(0,0))
v.zoomTo(viewer=(0,0))

# 3. Style the second cell (Ethanol on the right)
v.setStyle({'stick': {}}, viewer=(0,1))
v.zoomTo(viewer=(0,1))

# 4. Display the grid
v.show()
```
